## Document Loading and Preprocessing (RAG Pipeline)

### Objective
- Load a document (PDF/TXT/DOCX)
- Preprocess using LangChain
- Split into meaningful chunks (~500 tokens)
- Display representative chunks

```python
# 📦 Install required libraries
!pip install langchain langchain-community langchain-text-splitters pymupdf unstructured
```

```python
#  Import required modules
from langchain_community.document_loaders import PyMuPDFLoader, TextLoader, UnstructuredFileLoader
from langchain_text_splitters import RecursiveCharacterTextSplitter
```

```python
#  Upload your document (PDF/TXT/DOCX)
from google.colab import files
uploaded = files.upload()
```

```python
#  Load document (auto-detect type)

file_name = list(uploaded.keys())[0]

if file_name.endswith(".pdf"):
    loader = PyMuPDFLoader(file_name)
elif file_name.endswith(".txt"):
    loader = TextLoader(file_name)
else:
    loader = UnstructuredFileLoader(file_name)

documents = loader.load()

print(f"Total pages/documents loaded: {len(documents)}")
```

```python
#  Split document into chunks (~500 tokens)

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=500,
    chunk_overlap=50
)

chunks = text_splitter.split_documents(documents)

print(f"Total chunks created: {len(chunks)}")
```

```python
#  Display first 3 representative chunks

for i, chunk in enumerate(chunks[:3]):
    print(f"\n--- Chunk {i+1} ---\n")
    print(chunk.page_content)
```

#  Capstone Project Part 4 - Task 2
## Text Embedding and Vector Store Setup

### Objective
- Generate embeddings for text chunks
- Store in vector database (FAISS)
- Perform semantic search
- Retrieve top 3 relevant chunks

```python
#  Install additional dependencies
!pip install faiss-cpu sentence-transformers
```

```python

from langchain_community.embeddings import HuggingFaceEmbeddings
from langchain_community.vectorstores import FAISS
```

```python
#  Create Embedding Model

embedding_model = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)
```

```python
#  Create Vector Store (FAISS)

vector_db = FAISS.from_documents(
    documents=chunks,
    embedding=embedding_model
)

print("Vector database created successfully")
```

```python
#  Perform Semantic Search

query = "What are the key policies or important points discussed in the document?"

results = vector_db.similarity_search(query, k=3)
```

```python
#  Perform Semantic Search

query = "What are the key policies or important points discussed in the document?"

results = vector_db.similarity_search(query, k=3)
```

```python
#  Display Top 3 Results

print(f"\n Query: {query}\n")

for i, res in enumerate(results):
    print(f"\n--- Result {i+1} ---\n")
    print(res.page_content)
```

- Simulate multi-turn conversation

```python
from langchain_openai import ChatOpenAI
from langchain_community.vectorstores import FAISS

# ✅ Correct memory import
from langchain.memory.buffer import ConversationBufferMemory

# ✅ Correct chain import
from langchain.chains.conversational_retrieval.base import ConversationalRetrievalChain
```

```python
# 🤖 Initialize LLM

llm = ChatOpenAI(
    model="gpt-3.5-turbo",   # or gpt-4 if available
    temperature=0
)
```

```python
# 🧠 Add Memory

memory = ConversationBufferMemory(
    memory_key="chat_history",
    return_messages=True
)
```

```python
qa_chain = ConversationalRetrievalChain.from_llm(
    llm=llm,
    retriever=vector_db.as_retriever(),
    memory=memory
)
```

```python
# Turn 1
q1 = "Summarize the document"
r1 = qa_chain({"question": q1})
print("\nUser:", q1)
print("Assistant:", r1["answer"])

# Turn 2
q2 = "Explain it in simple terms"
r2 = qa_chain({"question": q2})
print("\nUser:", q2)
print("Assistant:", r2["answer"])

# Turn 3
q3 = "What are the key takeaways?"
r3 = qa_chain({"question": q3})
print("\nUser:", q3)
print("Assistant:", r3["answer"])
```

# Conversational Memory: Enhancing Context-Awareness and Continuity

Conversational memory was implemented using ConversationBufferMemory, which stores the full history of interactions between the user and the assistant.

### How Memory Improves Context-Awareness

- Each user query is not treated in isolation  
- Previous questions and responses are retained as chat history  
- The model uses this history to interpret follow-up queries correctly  

Example:
- User: "Summarize the document"  
- User: "Explain that in simple terms"  

The second query depends on the previous response. Memory allows the model to understand what "that" refers to.

---

### How Memory Enables Continuity

- Maintains a multi-turn conversation flow  
- Allows the assistant to:
  - Build on previous answers  
  - Avoid repetition  
  - Provide more refined responses  

Example flow:
1. Initial summary of the document  
2. Simplified explanation of the same content  
3. Extraction of key takeaways  

Each step builds logically on the previous one.

---

### Impact on RAG System

- Combines retrieved knowledge (RAG) with conversation history  
- Produces:
  - More relevant answers  
  - Better user experience  
  - More natural interaction  

---

### Key Benefit

Conversational memory transforms the system from a single-turn question-answer model into a context-aware assistant capable of maintaining coherent multi-step interactions.
